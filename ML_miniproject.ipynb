# Stanadardization
Here i apply feature scaling so that all parameters fit between 0 and 1 to save computational time and my gradient converge. since we deal with numbers between $10^1 - 10^3$

I selected min-max normalization given by and for each parameter in the dataset $\phi$ we rescale to $\phi'$
$$
\phi' = \frac{\phi - min(\phi)}{max(\phi) -min(\phi)}
$$


````python


def minmax(df, key):
  df[key] = (df[key] - df[key].min())/(df[key].max()- df[key].min())
  return df[key]

for item_keys in league_clean.keys():
  minmax(league_clean, item_keys)
````

Now lets build the model, first we split the dataset into a training and test set **WHY**
Using the sk.learn module we can randomly split the dataset into a test and training set, here split by 20% of the original data, same for the y outcomes.

````python
X_train, X_test, y_train, y_test = train_test_split(league_clean, match_outcome, test_size=0.2, random_state=50)
#convert the data into tensors
X_train = torch.FloatTensor(X_train.values)
X_test = torch.FloatTensor(X_test.values)
y_test = torch.FloatTensor(y_test.values)
y_train = torch.FloatTensor(y_train.values)
````

````python
class LOL_classification(torch.nn.Module):
    def __init__(self,shape,hidden,output):
      #create a linear regression model using multi-layer perceptron network
        super(LOL_classification, self).__init__()
        self.linear = torch.nn.Linear(shape, hidden)
        self.linear2 = torch.nn.Linear(hidden,hidden)

        self.linear3 = torch.nn.Linear(hidden, output)
        self.sigmoid = torch.nn.Sigmoid()

    def forward(self, x):
      x = self.linear(x)
      x = torch.relu(x)
      x = self.linear2(x)
      x = torch.relu(x)
      x = self.linear3(x)
      x = self.sigmoid(x)
      return x
````

The model above mathemitcally is
$$
y = \sigma(W_3^T(Relu(W_2^T(ReluW^T_1\vec{x} + b_1) + b_2) + b_3)
$$


Where $W_1$ = [37XN] dimenison linear matrix, $W_2$ = [NXN] and W_3 = [NX1] Where N is the dimension of the hidden layer, in my model i select N=2. For N=2 this leaves us with 85 parameters.
$$
((37\cdot 2) + 2) + ((2 \cdot2) +2) + ((2 \cdot 1) + 1)) = 85
$$
We set the width of the last tensor to 1 to return a sigular output for binary classification. We also pass the Relu function to avoid the vanishing gradient problem.

````python
#evaluate accuracy - got for my assignment 4 and chatgpt
def accuracy(model, data, target):
    # Set the model to evaluation mode
    model.eval()

    # Disable gradient computation
    with torch.no_grad():
        # Forward pass to get the predictions
        output = model(data).squeeze(1)

        # Get the index of the max log-probability (predicted class)
        pred = np.around(output.numpy())

        # Convert predictions and target to numpy arrays
        target = target.numpy().flatten()

        # Round the predictions using np.around

        # Calculate the fraction of matches
        matches = np.sum(np.equal(pred, target))
        total = len(target)
        acc = matches / total

    return acc
````

````python
#number of parameters in the dataset
shape = len(league_clean.columns)

model = LOL_classification(shape=shape,hidden=2,output=1)
criterion = torch.nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr =0.01)
````

````python
# Training loop  chatgpt

epochs = 10000
loss_history = []
print("Training Loop")
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    y_pred = model(X_train)
    loss = criterion(y_pred.squeeze(), y_train)
    loss.backward()
    loss_history.append(loss.detach().numpy())
    optimizer.step()
    if epoch% 1000 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# Evaluation
acc = accuracy(model,X_train, y_train)
print(f"Accuracy for Training model: ",{acc*100},'%')
````

**Repeat for the test model**

````python
loss_history_test = []
print("Test Loop")
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    y_pred = model(X_test)
    loss = criterion(y_pred.squeeze(), y_test)
    loss.backward()
    #append loss for plotting covnert to numpy object
    loss_history_test.append(loss.detach().numpy())
    optimizer.step()
    if epoch% 1000 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
acc_test = accuracy(model,X_test, y_test)
print(f"Accuracy for Training model: ",{acc*100},'%')
````
