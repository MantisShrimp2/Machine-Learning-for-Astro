Why i remove some parameters:

# 1) Monsters
LoL has many monsters such as dragons Rift Herald, Baron Nasor and void grubs that provide additional benefits to players such as gold, power and experience.
Teams who obtain these monsters gain significant advantage throughout the game. However since i am interested in if a parameter offers an advantage or not, I omit subcategories of dragons (Cloud, Infernal, Mountain, Ocean, Hextech) and keep the total number of dragons as a parameter. This way i can observe how monsters and dragons as a whole provide a team. I do this in general for the other parameters i omit.

**Keep the primary parameter in the model and omit sub-categories of the parameter.**

One important note, this dataset only collects data for the first 15 minutes of each game, Inhibtiors are late game objectives (>20 minutes) so it is unlikely many are in the dataset.

Now lets build the model, first we split the dataset into a training and test set **WHY**
Using the sk.learn module we can randomly split the dataset into a test and training set, here split by 20% of the original data, same for the y outcomes.

````python
X_train, X_test, y_train, y_test = train_test_split(league_clean, match_outcome, test_size=0.2, random_state=50)
#convert the data into tensors
X_train = torch.FloatTensor(X_train.values)
X_test = torch.FloatTensor(X_test.values)
y_test = torch.FloatTensor(y_test.values)
y_train = torch.FloatTensor(y_train.values)
````

````python
class LOL_classification(torch.nn.Module):
    def __init__(self,shape):
        super(LOL_classification, self).__init__()
        self.linear = torch.nn.Linear(shape, 1)
        self.sigmoid = torch.nn.Sigmoid()

    def forward(self, x):
        return self.sigmoid(self.linear(x))


````

````python
#evaluate accuracy - got for my assignment 4 and chatgpt
def accuracy(model, data, target):
    # Set the model to evaluation mode
    model.eval()

    # Disable gradient computation
    with torch.no_grad():
        # Forward pass to get the predictions
        output = model(data).squeeze(1)

        # Get the index of the max log-probability (predicted class)
        pred = np.around(output.numpy())

        # Convert predictions and target to numpy arrays
        target = target.numpy().flatten()

        # Round the predictions using np.around

        # Calculate the fraction of matches
        matches = np.sum(np.equal(pred, target))
        total = len(target)
        acc = matches / total

    return acc
````

````python
shape = len(league_clean.columns)

model = LOL_classification(shape=shape)
criterion = criterion = torch.nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
````

````python
# Training loop  chatgpt

epochs = 10000
loss_history = []
print("Training Loop")
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    y_pred = model(X_train)
    loss = criterion(y_pred.squeeze(), y_train)
    loss.backward()
    loss_history.append(loss.detach().numpy())
    optimizer.step()
    if epoch% 1000 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# Evaluation
acc = accuracy(model,X_train, y_train)
print(f"Accuracy for Training model: ",{acc*100},'%')
````

**Repeat for the test model**

````python
loss_history_test = []
print("Test Loop")
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    y_pred = model(X_test)
    loss = criterion(y_pred.squeeze(), y_test)
    loss.backward()
    #append loss for plotting covnert to numpy object
    loss_history_test.append(loss.detach().numpy())
    optimizer.step()
    if epoch% 1000 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
acc_test = accuracy(model,X_test, y_test)
print(f"Accuracy for Training model: ",{acc*100},'%')
````

````python
n_epochs = np.linspace(1,epochs,num=epochs)
plt.plot(n_epochs, loss_history,label='Training data')
plt.plot(n_epochs, loss_history_test, label='Test data')
plt.title("Loss History over Epochs")
plt.grid()
plt.xlabel("Epoch")
plt.legend()
plt.ylabel("Loss History")
plt.show()
````

````python
model.eval()
with torch.no_grad():
    # Predictions for training data
    train_inputs = torch.FloatTensor(X_train)
    train_outputs = model(train_inputs)
    train_predicted = (train_outputs.squeeze())

    # Predictions for testing data
    test_inputs = torch.FloatTensor(X_test)
    test_outputs = model(test_inputs)
    test_predicted = (test_outputs.squeeze())


# Plotting
plt.figure(figsize=(12, 6))

# Plot training data
plt.subplot(1, 2, 1)
plt.scatter(range(len(y_train)), y_train, label='Actual', color='blue', alpha=0.6)
plt.scatter(range(len(train_predicted)), train_predicted, label='Predicted', color='red', marker='x', alpha=0.6)
plt.xlabel('Game Number')
plt.ylabel('Win Outcome (1: Win, 0: Loss)')
plt.title('Training Data: Actual vs Predicted Win Outcome')
plt.legend()

# Plot testing data
plt.subplot(1, 2, 2)
plt.scatter(range(len(y_test)), y_test, label='Actual', color='blue', alpha=0.6)
plt.scatter(range(len(test_predicted)), test_predicted, label='Predicted', color='red', marker='x', alpha=0.6)
plt.xlabel('Game Number')
plt.ylabel('Win Outcome (1: Win, 0: Loss)')
plt.title('Testing Data: Actual vs Predicted Win Outcome')
plt.legend()

plt.tight_layout()
plt.show()
````
