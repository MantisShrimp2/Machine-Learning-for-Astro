a. Implement a sampling algorithm that samples $(X, Y)$ pairs.

```python
def return_random_sample_joint():
    '''Sample a probability from the table above
    Select a random number with px between 0 and 1
    assign a sample pair to putput based on  the random number

    return a sample pair
    '''
    px = random()
    if px < 0.1:
        return [0,1]
    elif px < 0.2:
        return [1,0]
    elif px < 0.3:
        return [1,1]
    else:
        return [0,0]
x = return_random_sample_joint()
```

### 2.2 Marginal normal distributions (3 points)

*Topics: multivariate normal & marginal distributions*

Consider a multivariate Gaussian of the form
$$P(\mathbf{x}) = \mathcal{N}(\mathbf{x}|\mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{(2\pi)^{D/2} |\mathbf{\Sigma}|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x}-\boldsymbol{\mu})^{T} \mathbf{\Sigma}^{-1}(\mathbf{x}-\boldsymbol{\mu})\right) \, ,$$

where $\mathbf{x}$ and $\boldsymbol{\mu}$ are vectors with $D$ entries and $\mathbf{\Sigma}$ is a $D \times D$ covariance matrix. $|\mathbf{\Sigma}|$ is the determinant of this matrix.

#### a (2 points)
Consider the case where $D = 2$. Use the following three sets of ($\boldsymbol{\mu}$, $\mathbf{\Sigma}$) to create 2D contour- and 3D surface plots of the corresponding PDFs (with arbitrary contour levels)

$$ \boldsymbol{\mu_1} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_1} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} \, ,$$

$$ \boldsymbol{\mu_2} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_2} = \begin{pmatrix} 1 & -0.5 \\ -0.5 & 1 \end{pmatrix} \, ,$$

$$ \boldsymbol{\mu_3} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_3} = \begin{pmatrix} 1 & 0.7 \\ 0.7 & 1 \end{pmatrix} \, .$$

Implement the multivariate Gaussian yourself using numpy matrix multiplication.

```python
import numpy as np
import matplotlib.pyplot as plt

def mvgauss(x, mu, sigma):
    det = np.linalg.det(sigma) # calculate the determinate of the matrix
    inv = np.linalg.inv(sigma)
    D = 2
    #the coefficent outside the expoential
    coeff = 1/((det**0.50))*(2*np.pi)**(D/2)
    #calculate the matriricies
    mtrx_A = np.dot(np.transpose(x-mu),inv)
    mtrx_B = np.dot(mtrx_A, (x-mu))
    
    return coeff*np.exp(-0.5*mtrx_B)

#define each matrix
mu1 = np.array([5,5])
sigma1 = np.array([[1,0],
                  [0,1]])
sigma2 = np.array([[1,-0.5],
                  [-0.5,1]])

sigma3 = np.array([[1,0.7],
                  [0.7,1]])

#x = np.array([0,1])
```

```python
from mpl_toolkits.mplot3d import Axes3D


def plot_contour(mu, sigma, sigma_label):
    # Generate grid of points
    x = np.linspace(-10,10)
    y = np.linspace(-10,10)
    X, Y = np.meshgrid(x, y)
   # points = np.vstack([X.ravel(), Y.ravel()])

    
    # Compute PDF values at grid points
    Z = [mvgauss([x_val,y_val], mu, sigma) for x_val,y_val in zip(np.ravel(X),np.ravel(Y))]
    Z = np.array(Z) # convert to numpy array
         
    Z = Z.reshape(X.shape)

    # Plot contour
    plt.figure()
    plt.contour(X, Y, Z)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('2D Multivariate Contour Plot for '+sigma_label)
    plt.show()

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('PDF')
    ax.set_title('3D Multivariate Contour Plot for '+sigma_label)
    plt.show()
    return None



# Plot 2D contour
plot_contour(mu1, sigma1, "Sigma 1")
```
